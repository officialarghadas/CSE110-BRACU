```python
#TASK 02
def fibonacci(num):
  first=0
  second=1
  if num==0:
    print(first, end=" ")
  elif num>1:
    print(first, end=" ")
    print(second, end=" ")
    while True:
      third=first+second
      if third>num:
        break
      print(third, end=" ")
      first=second
      second=third
  else:
    pass
fibonacci(9)
```

```python
#task 7

def show_palindrome(num):
  str1=""
  for i in range(1,num+1):
    str1+=str(i)
  for i in range(num-1, 0, -1):
    str1+=str(i)
  return str1
print(show_palindrome(5))

```

```python
#task 13

def numfuction(opt,num1,num2):
    if opt =='+':
        x=num1+num2
        return x
    elif opt=='-':
        x=num1-num2
        return x
    elif opt=='*':
        x=num1*num2
        return x
    elif opt=='/':
        x=num1/num2
        return x
opt=input('Enter oparetor: ')
num1=float(input('Enter 1st number: '))
num2=float(input('Enter 2nd number: '))

print(numfuction(opt,num1,num2))
```

```python
#TASK 14

def cuttingcharacters(sentence,position):
    temp=''
    counter=''
    for p in range(0,len(sentence)):
        if p!=0 and p%position==0:
            temp+=sentence[p]
        else:
            counter+=sentence[p]

    return counter+temp
sentence=str(input('Enter sentence: '))
position=int(input('Enter position: '))
print(cuttingcharacters(sentence,position))
```
